# Аггрегирование данных

```python
import pandas as pd
df = pd.read_csv('../datasets/real_estate_data.csv', sep = '\t')
```

```python
df['balcony'] = df['balcony'].fillna(0)
```

```python
# Посчитать кол-во недвижимости по каждому городу
df.groupby('locality_name').agg({'total_images':'count'})
```

```python
# посчитать кол-во недвидимотсти по ков-ву комнат
df.groupby('rooms').agg({'total_images': 'count'}) # в agg выибраем самый наполненный стоблец
```

```python
# посчитать кол-во недвижимости по наличию балконов
df.groupby('balcony').agg({'total_images': 'count'})
```

```python
# посчитать среднюю высоту потолков по каждому городу
df.groupby('locality_name').agg({'ceiling_height': 'mean'})
```

```python
# посчитать среднюю высоту потолков по кол-ву комнат
df.groupby('rooms').agg({'ceiling_height': 'mean'})
```

# Использование метода sort_values вместе с group_by и agg

```python
# выведите города с самой высокой суммой по всем недвижимостям
# reset_index() - чтобы добавить индексы слева

df.groupby('locality_name').agg({'last_price': 'sum'}).reset_index().sort_values(by='last_price', ascending=True)
```

```python
# выведите города, с наибольшим кол-вом недивжимости

df.groupby('locality_name').agg({'total_images': 'count'}).reset_index().sort_values(by='total_images', ascending=False)
```

```python
# выведите тип недвижимости по комнатам, объеявления которых больше всего

df.groupby('rooms').agg({'total_images': 'count'}).reset_index().sort_values(by='total_images', ascending=False)
```

```python
# Выведите тип недвижимости по комнатам, где имеется высокая средняя жил-площадь

df.groupby('rooms').agg({'living_area': 'mean'}).reset_index().sort_values(by='living_area', ascending=False)
```
