## 1. Посчитайте средний возраст пассажиров сгруппированных по классам.



```python
average_age_by_class = df1.groupby('pclass')['age'].mean()
print(average_age_by_class)
```

## 2. Выявите какого возраста пассажиры выжили больше всего



```python
# Отфильтровать данные для выживших пассажиров и найти моду возраста
most_survived_age = df1[df1['survived'] == 1]['age'].mode()[0]
print("Наибольшее количество выживших пассажиров было возраста:", most_survived_age)
```

## 3. Выявите какие возраста в основном были рассредоточены по классам, а также среднюю плату за билеты среди них



```python
# Группировка данных по классу и возрасту, и подсчет средней платы за билеты
average_fare_by_age_class = df1.groupby(['pclass', 'age'])['fare'].mean()
average_fare_by_age_class
```

```python
# Найти возраста, которые в основном рассредоточены по классам
most_varied_ages = average_fare_by_age_class.groupby('age').std().idxmax()
most_varied_ages
```

```python
# Средняя плата за билеты для найденных возрастов
average_fare_for_most_varied_ages = average_fare_by_age_class.loc(axis=0)[:, most_varied_ages].mean()
average_fare_for_most_varied_ages 
```

```python
print("Возраста, в основном рассредоточенные по классам:", most_varied_ages)
print("Средняя плата за билеты для этих возрастов:", average_fare_for_most_varied_ages)
```

## 4. Выявите в каком порту было больше всего пассажиров до 40 лет

```python
# Отфильтровать данные для пассажиров моложе 40 лет
passengers_under_40 = df1[df1['age'] < 40]
passengers_under_40
```

```python
# Подсчитать количество пассажиров в каждом порту
passengers_by_port = passengers_under_40['embarked'].value_counts()
passengers_by_port
```

```python
# Найти порт с наибольшим количеством пассажиров до 40 лет
most_passengers_port = passengers_by_port.idxmax()
most_passengers_port
```

```python
print("Порт, в котором было больше всего пассажиров до 40 лет:", most_passengers_port)
```

## 5. Выявите самый популярный класс для пассажиров женского пола и среднюю цену поездки по ним

```python
# Отфильтровать данные для пассажиров женского пола
female_passengers = df1[df1['sex'] == 'female']
female_passengers
```

```python
# Подсчитать количество пассажиров в каждом классе
female_passengers_by_class = female_passengers['pclass'].value_counts()
female_passengers_by_class
```

```python
# Найти самый популярный класс для пассажиров женского пола
most_popular_class_female = female_passengers_by_class.idxmax()
most_popular_class_female 
```

```python
# Вычислить среднюю цену билета для самого популярного класса
average_fare_most_popular_class = female_passengers[female_passengers['pclass'] == most_popular_class_female]['fare'].mean()
average_fare_most_popular_class 
```

```python
print("Самый популярный класс для пассажиров женского пола:", most_popular_class_female)
print("Средняя цена поездки для этого класса:", average_fare_most_popular_class)
```

## 6. Разделив пассажиров по полам, выявите какие пассажиры имеют большую сумму платы за поездку.

```python
# Сгруппировать данные по полу и вычислить сумму платы за поездку для каждой группы
total_fare_by_gender = df1.groupby('sex')['fare'].sum()
total_fare_by_gender
```

```python
# Определить, у какой группы пассажиров сумма платы за поездку больше
gender_with_highest_total_fare = total_fare_by_gender.idxmax()
highest_total_fare = total_fare_by_gender.max()
print("Пассажиры с", gender_with_highest_total_fare, "пола имеют большую сумму платы за поездку:", highest_total_fare)
```

## 7. Теперь посчитайте сколько было заработано денег в целом



```python
# Посчитать общую сумму заработанных денег
total_earnings = df1['fare'].sum()
total_earnings
print("Общая сумма заработанных денег от продажи билетов:", total_earnings)
```

## 8. Предположим, что компания решила выплатить компенсацию, пострадавшим семьям. Посчитайте сколько компания должна выплатить родственникам погибших на титанике, если компания обязана выплатить 100тыс долларов за каждого погибшего человека.



```python
# Подсчитать количество погибших пассажиров
deceased_passengers = df1[df1['survived'] == 0]
total_deceased_passengers = deceased_passengers.shape[0]
deceased_passengers
```

```python
total_deceased_passengers
```

```python
# Сумма компенсации за каждого погибшего человека
compensation_per_deceased = 100000  # в долларах
compensation_per_deceased
```

```python
# Вычислить общую сумму компенсации
total_compensation = total_deceased_passengers * compensation_per_deceased
print("Общая сумма компенсации для родственников погибших на Титанике: $", total_compensation)
```

## 9. Теперь соберите вашу информацию и опишите вывод проделанного исследования. А также добавьте сумму убытка, которую можно посчитать, если вычесть общую сумму заработка и расходы на компенсации родственникам.



```python
# Общая сумма убытка
total_loss = total_earnings - total_compensation
total_loss
```
