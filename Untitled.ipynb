```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Пример данных (замените этот блок вашим реальным DataFrame)
data = {'City': ['Город1', 'Город2', 'Город1', 'Город3', 'Город2', 'Город3'],
        'Price': [500000, 600000, 550000, 700000, 650000, 750000]}
df = pd.DataFrame(data)

# Создаем график распределения цен по городам
plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='Price', data=df)

# Добавляем заголовок и метки осей
plt.title('Распределение цен на недвижимость по городам')
plt.xlabel('Город')
plt.ylabel('Цена')

# Показываем график
plt.show()
```
